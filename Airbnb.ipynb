```python
import pandas as pd

# Assuming df is your DataFrame with 300 columns

# List of 50 desired column names
desired_columns = ['_id', 'name', 'space', 'description', 'property_type','room_type', 'beds', 'bed_type', 'minimum_nights', 'maximum_nights',
                    'cancellation_policy', 'bedrooms', 'accommodates', 'bathrooms', 'price', 'weekly_price',
                    'monthly_price', 'cleaning_fee', 'security_deposit', 'extra_people', 'images.picture_url',
                    'host.host_url', 'host.host_id', 'host.host_location', 'host.host_response_time',
                    'host.host_name',
                    'host.host_identity_verified',
                    'address.street', 'address.suburb', 'address.government_area', 'address.market',
                    'address.country', 'address.country_code', 'address.location.type',
                    'address.location.coordinates[0]', 'address.location.coordinates[1]',
                    'address.location.is_location_exact',
                    'availability.availability_30',
                    'availability.availability_60', 'availability.availability_90',
                    'availability.availability_365', 'number_of_reviews','review_scores.review_scores_accuracy',
                    'review_scores.review_scores_cleanliness', 'review_scores.review_scores_checkin',
                    'review_scores.review_scores_communication', 'review_scores.review_scores_location', 'review_scores.review_scores_value',
                    'review_scores.review_scores_rating',
                   'amenities[0]', 'amenities[0]', 'amenities[1]',
                    'amenities[2]', 'amenities[3]', 'amenities[4]',
                    'guests_included'
                    ]

# Select the desired columns from the DataFrame
df = df[desired_columns]

# Display the first few rows of the selected DataFrame
df.head()
```

```python

# Specify the columns you want to clean (remove special characters +, @)
columns_to_clean = ['host.host_identity_verified','space', 'description']

# Remove special characters (+, @) from the specified columns
df[columns_to_clean] = df[columns_to_clean].replace({r'[+@]': ''}, regex=True)

# Replace underscores with spaces in the 'cancellation_policy' column
df['cancellation_policy'] = df['cancellation_policy'].str.replace('_', ' ')

# Display the DataFrame after cleaning
print("\nAfter cleaning:")
print(df)
print(df["cancellation_policy"])
```

```python
# Identifying categorical columns
categorical_columns = df.select_dtypes(include=['object', 'category']).columns.tolist()

# Identifying numerical columns
numerical_columns = df.select_dtypes(include=['int64', 'float64']).columns.tolist()

print("Categorical Columns:", categorical_columns)
print("Numerical Columns:", numerical_columns)

```

```python
# Step 1: List of numerical columns (same as before)
numerical_columns = ['beds', 'minimum_nights', 'maximum_nights', 'bedrooms', 'accommodates',
                     'bathrooms', 'price', 'weekly_price', 'monthly_price', 'cleaning_fee',
                     'security_deposit', 'extra_people', 'host.host_id',
                     'address.location.coordinates[0]', 'address.location.coordinates[1]',
                     'availability.availability_30', 'availability.availability_60',
                     'availability.availability_90', 'availability.availability_365',
                     'number_of_reviews', 'review_scores.review_scores_accuracy',
                     'review_scores.review_scores_cleanliness', 'review_scores.review_scores_checkin',
                     'review_scores.review_scores_communication', 'review_scores.review_scores_location',
                     'review_scores.review_scores_value', 'review_scores.review_scores_rating',
                     'guests_included']

# Step 2: Filter the DataFrame to include only the numerical columns
numerical_df = df[numerical_columns]

# Step 3: Fill NaN values for each numerical column
for col in numerical_df:
    if df[col].skew() < -0.5 or df[col].skew() > 0.5:  # Check if it's skewed
        df[col].fillna(df[col].median(), inplace=True)
    else:
        df[col].fillna(df[col].mean(), inplace=True)

# Step 4: Check if any column still contains NaN values
null_columns = numerical_df.isnull().sum()

# Filter and print columns with remaining null values
remaining_null_columns = null_columns[null_columns > 0]
if remaining_null_columns.empty:
    print("All missing values have been filled.")
else:
    print("The following columns still contain missing values:")
    print(remaining_null_columns)
```

```python
numerical_df["bedrooms"] = numerical_df["bedrooms"].astype(int)
numerical_df["beds"] = numerical_df["beds"].astype(int)
numerical_df["bathrooms"] = numerical_df["bathrooms"].astype(str).astype(float).astype(int)
numerical_df["price"] = numerical_df["price"].astype(int)
numerical_df["extra_people"] = numerical_df["extra_people"].astype(int)
numerical_df["cleaning_fee"] = numerical_df["cleaning_fee"].astype(int)
numerical_df["security_deposit"] = numerical_df["security_deposit"].astype(int)
numerical_df["weekly_price"] = numerical_df["weekly_price"].astype(int)
numerical_df["monthly_price"] = numerical_df["monthly_price"].astype(int)
numerical_df["review_scores.review_scores_accuracy"] = numerical_df["review_scores.review_scores_accuracy"].astype(int)
numerical_df["review_scores.review_scores_cleanliness"] = numerical_df["review_scores.review_scores_cleanliness"].astype(int)
numerical_df["review_scores.review_scores_checkin"] = numerical_df["review_scores.review_scores_checkin"].astype(int)
numerical_df["review_scores.review_scores_communication"] = numerical_df["review_scores.review_scores_communication"].astype(int)
numerical_df["review_scores.review_scores_location"] = numerical_df["review_scores.review_scores_location"].astype(int)
numerical_df["review_scores.review_scores_value"] = numerical_df["review_scores.review_scores_value"].astype(int)
numerical_df["review_scores.review_scores_rating"] = numerical_df["review_scores.review_scores_rating"].astype(int)
```

```python
df1 = pd.DataFrame()  # Create an empty DataFrame to store the modified columns
for col in ['name', 'space', 'description', 'property_type', 'room_type', 'bed_type', 'cancellation_policy',
            'images.picture_url', 'address.street', 'address.suburb', 'address.government_area', 'address.market',
            'address.country', 'address.country_code', 'address.location.type',
            'host.host_url', 'host.host_location', 'host.host_response_time',
            'host.host_name', "host.host_identity_verified"]:

    df1[col] = df[col].fillna('Unknown')  # Do not use inplace=True, return the modified column and assign it to df1

print(df1)
```

```python
# Step 1: Apply fillna based on data type for the entire DataFrame
df1 = df1.apply(lambda col: col.fillna('Unknown') if col.dtype == 'object' else col.fillna(0))
print(df1)
# Now your DataFrame has 'Unknown' for NaN in categorical columns and 0 for NaN in numerical columns.

```
